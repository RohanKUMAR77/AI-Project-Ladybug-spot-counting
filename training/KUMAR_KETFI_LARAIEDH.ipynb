#### Fonction pour récupérer l'indice du 2ème maximum (qui correspond à la couleur du corps de la coccinelle)

```python
def scnd_max_index(liste):
    # Supposons que la liste ait au moins deux éléments
    maxi = max(liste[0], liste[1])
    max_index = liste.index(maxi)
    scnd_max = min(liste[0], liste[1])
    scnd_max_index = liste.index(scnd_max)

    for i in range(2, len(liste)):
        if liste[i] > maxi:
            scnd_max = maxi
            scnd_max_index = max_index
            maxi = liste[i]
            max_index = i
        elif liste[i] > scnd_max:
            scnd_max = liste[i]
            scnd_max_index = i

    return scnd_max_index
```

#### Fonctions pour calculer le point moyen et la distance entre 2 points 

```python
def mean_point(points):
    n = len(points)
    somme_x = sum(point[0] for point in points)
    somme_y = sum(point[1] for point in points)
    somme_z = sum(point[2] for point in points)
    mean_point = (somme_x/n, somme_y/n, somme_z/n)
    return mean_point

def distance(x, y):
    distance = np.sqrt((y[0] - x[0])**2 + (y[1] - x[1])**2 + (y[2] - x[2])**2)
    return distance
```

### La focntion get_color permet de récupérer le nombre de couleur qu'il y a sur l'image segmentée et aussi de récupérer la couleur du corps de la coccinelle

```python
def get_color(images, seg_images):
    #On initailise la liste des couleurs et la liste du nombre de couleurs
    ladybug_tint = []
    seg_colors_count = []

    for k in range (len(images)):
        sys.stdout.write("\rImages traitées : {}/{}".format(k+1, len(images)))
        ## Prenons l'image segemtée 
        img = Image.open(seg_images[k])

        ###Ensuite on construit l'array qui correspond aux couleurs de chaque pixels 
        img_array = np.asarray(img)

        ## Ensuite le but est de récupérer les pixels qui correspondent à la couleur du corps de la coccinelle
        ## On a de la chance dans toues le scas ceci correspond à la 2ème couleur la plus dominante 

        # Commandes qui permettent de compter pour chaque couleur le nombre de pixels    
        pixels = img_array.reshape(-1, img_array.shape[-1])
        # On compte le nombre de pixels pour chaque couleur
        color_counts = Counter(map(tuple, pixels))
        # Liste des couleurs dans l'image associée au comptage effectué
        color_list = list(color_counts.keys())



        # On peut ajouter le nombre de couleurs dans l'image
        seg_colors_count.append(len(color_list))
        # On récupère la liste du nombre de pixels pour chaque couleur
        pixel_counts = list(color_counts.values())
        
        #On récupère l'indice de la couleur du corps de la coccinelle qui correspond à la 2ème couleur la plus présente
        #On utilise la fonction scnd_max_index
        index_color = scnd_max_index(pixel_counts)
        #Du coup on peut maintenant récupérer la couleur gray scale du corps de la coccinelle
        body_color_gray = color_list[index_color]

        #On récupère les coordonnées de tous les pixels correspondant au corps
        pixel_body = []
        for i in range(img_array.shape[0]):
            for j in range(img_array.shape[1]):
                if np.array_equal(img_array[i][j], body_color_gray):
                    pixel_body.append([i,j])

        #L'idée maintenant serait de prendre dans l'array image toutes les couleurs correpsondent au coordonnées du corps 
        #de la coccinelle
        #On ouvre l'image
        img2 = Image.open(images[k])
        img2_array = np.asarray(img2)

        body_color_list = []
        for k in pixel_body:
            body_color_list.append(list(img2_array[k[0]][k[1]]))

        #Calculons maintenant la couleur moyenne

        mean_color = list(mean_point(body_color_list))
        color_dic = {"black": [0,0,0], "red": [204,51,51], "orange": [204,102,0], "darkred": [100, 0, 0]}

        min_dist = distance([0,0,0], mean_color)
        ladybug_color = "black"

        for i,j in color_dic.items():
            dist = distance(j, mean_color)
            if (dist<min_dist):
                min_dist = dist
                ladybug_color = i

        ladybug_tint.append(ladybug_color)
        sys.stdout.flush()
    return seg_colors_count, ladybug_tint
```

#### Pour récupérer le nombre de clusters dans le corps de la coccinelle : 
    On retire les 3 couleurs les plus fortes de l'array de l'image segmentée qui correspondent : 
        - Au fond
        - Au corps de la coccinelle
        - La tête de la coccinelle
    Ensuite il nous reste que les points et certaines couleurs présentent dans la tête comme les yeux
    On se débarasse de ces détails restant en vérifiant que chaque coordonnée de l'array est bien à l'extérieur de la tête
    
    Pour vérifier cela on prend un quadrilatère qui correspond au minimum et maximum sur chaque axe
        

```python
# ALgorithme pour compter le nombre de cluster dans le corps de la coccinelle (ça devrait se rapprocher du nombre de points)
```

```python
# Quelques fonctions utiles : 

def top3_index(L):
    indexes = sorted(range(len(L)), key=lambda i: L[i], reverse=True)[:3]
    return indexes


def remove_index(L, index):
    res = []
    for k in range (len(L)):
        if k not in index: 
            res.append(L[k])
    return res
```

### Fonction pour compter le nombre de clusters 

```python
def body_cluster(images, seg_images):
    body_clusters = []
    for k in range (len(images)):

        sys.stdout.write("\rImages traitées : {}/{}".format(k+1, len(images)))
        #On lit l'image segmentée
        img = Image.open(seg_images[k])
        img_array = np.asarray(img)
        # Commandes qui permettent de compter pour chaque couleur le nombre de pixels    
        pixels = img_array.reshape(-1, img_array.shape[-1])
        # On compte le nombre de pixels pour chaque couleur
        color_counts = Counter(map(tuple, pixels))
        #Le résultat est un dictionnaire avec les couleurs qui sont les clés et le nombre de pixels les valeurs des clés


        # Liste des couleurs dans l'image associée au comptage effectué
        color_list = [np.array(k, dtype=int) for k in list(color_counts.keys())]
        # De même pour la liste du nombre de pixels pour chaque couleur
        pixel_counts = list(color_counts.values())
        # Maintenant on récupère les indices des couleurs avec le plus de Pixels grâce à notre fonction top3_index
        indexes = top3_index(pixel_counts)

        #### PARTIE OU ON EXTRAIT LES POINTS DE LA TETE

        #La 3ème valeur correspond à la couleur de la tête
        head_color = color_list[indexes[2]]
        head_points = []
        #print(type(head_color))
        #print(type(img_array[0][0]))
        for i in range(len(img_array)):
            for j in range(len(img_array[0])):
                if np.array_equal(img_array[i][j], head_color):
                    head_points.append([i,j])

        xmin, xmax = head_points[0][0], head_points[0][0]
        ymin, ymax = head_points[0][1], head_points[0][1]
        for point in head_points: 
            if point[0]<xmin:
                xmin = point[0]
            if point[0]>xmax:
                xmax = point[0]
            if point[1]<ymin:
                ymin = point[1]
            if point[1]>ymax:
                ymax = point[1]

        ########## PARTIE ON COMPTE LES NB DE POINTS
        # On retire ces couleurs dans notre liste de couleurs grâce à la fonction remove_index
        color_list = remove_index(color_list, indexes)
        # Maintenant dans color_list on a que les couleurs des points et d'autres détails comme les yeux


        # On parcourt l'array de l'image et si on tombe sur une couleur qui correspond à celle d'un point 
        # On ajoute les coordonnées dans la liste des coordonnées
        spots_pixels = []
        for i in range (img_array.shape[0]):
            for j in range (img_array.shape[1]):
                if np.any(np.all(img_array[i, j] == color_list, axis=1)):
                    if (i<xmin)or(i>xmax)or(j<ymin)or(j>ymax):
                        spots_pixels.append([i, j])

        # Maintenant il nous reste plus qu'à appliquer la méthode de clustering hiérarchique
        # Avec la distance minimale entre les clusters
        # Pour trouver le nombre de points + yeux de la coccinelle

        # Initialisation de l'algorithme de regroupement hiérarchique agglomératif avec la distance minimale
        clustering = AgglomerativeClustering(n_clusters=None, linkage='single', distance_threshold = 4).fit(spots_pixels)
        # Récupération du nombre de clusters
        n_spots = clustering.n_clusters_
        body_clusters.append(n_spots)
        sys.stdout.flush()
    return body_clusters
```

### On entraine sur 95 % du dataframe pour prédire l'espèce

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import pickle

#On a le vecteur X qui sont nos homemade features 
X = df[['seg_colors_count', 'body_red', 'body_green', 'body_clusters']]

#Le vecteur y qui est le label
y = df['Label']

#On a donc les vecteurs suivant pour l'entrainement et le test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05, random_state=42)


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

#On va utiliser la regression logistique pour prédire le Label
model = LogisticRegression()
model.fit(X_train_scaled, y_train)

#On applique la prédiction
y_pred = model.predict(X_test_scaled)

# On évalue l'accuracy de la phase d'entrainement
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

with open('model.pkl', 'wb') as file:
    pickle.dump(model, file)
```

#### Avec la PCA on identifie 5 clusters plus ou moins

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import linkage, fcluster
from scipy.spatial.distance import pdist

#On va utiliser un clustering hiérarchique pour proposer un cluster
X = df[['seg_colors_count', 'body_red', 'body_green', 'body_clusters']]


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


distances = pdist(X_scaled)

# Step 6: Perform hierarchical clustering
hierarchy = linkage(distances, method='single')

# Step 7: Assign cluster labels based on the minimum distance threshold
min_distance = 0.7
cluster_labels = fcluster(hierarchy, min_distance, criterion='distance')

# Step 8: Add the cluster labels to the DataFrame
df['Cluster'] = cluster_labels
```
